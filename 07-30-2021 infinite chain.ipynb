```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
theta_range = np.linspace(-math.pi,math.pi,360)
fraction_range = np.linspace(.1,.9,3)
chain_length = 1000
```

```python
# recursive function for calculating target link position
def calculatePosition(chain, theta, link):
    current_angle = theta * (link + 1)
    if link == 0:
        return np.array([1 + chain[link] * math.cos(current_angle), chain[link] * math.sin(current_angle)])
    else:
        return calculatePosition(chain, theta, link - 1) \
                + np.array([chain[link] * math.cos(current_angle), chain[link] * math.sin(current_angle)])
    
```

```python
fig, ax = plt.subplots(figsize = (12,12))
ax.set_aspect('equal')
for f in fraction_range:
    chain = [(f)**(i+1) for i in range(chain_length)]
    x = []
    y = []
    for theta in theta_range:
        x.append(calculatePosition(chain, theta, len(chain) - 1)[0])
        y.append(calculatePosition(chain, theta, len(chain) - 1)[1])
    plt.scatter(x, y, alpha = .6, s = 15)
plt.legend(fraction_range)
plt.title('End point Position from [-pi,pi] colored by chain link fraction (f)')
plt.savefig('chain_graph');
```
