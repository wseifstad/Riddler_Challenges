```python
import itertools
```

```python
teams = [1]*4 + [0]*4
```

```python
# perms = [x[:7] for x in set(itertools.permutations(teams))]
# perms = [x[:7] for x in itertools.permutations(teams)] # keep all, not just unique perms
perms = [i for i in list(itertools.product('01', repeat=7))]
```

```python
len(perms)
```

```python
outcomes = []
for perm in perms:
    team1wins = 0
    team2wins = 0
    for i,j in enumerate(perm):
        if int(j) > 0: 
            team1wins += 1
        else: 
            team2wins += 1
        if i > 2 and (team1wins == 4 or team2wins == 4): 
            outcomes.append(perm[:i+1])
            break
outcomes = list(outcomes)
```

```python
len(outcomes) #outcomes
```

```python
lens = 0
for i in outcomes:
    lens += len(i)
expected_number_of_games = (lens / len(perms))
expected_number_of_games
```

```python
guesses = set(perms)
```

```python
best_result = 0
best_guess = tuple()
for guess in guesses:
    result = 0
    incorrect_outcomes = []
    for outcome in outcomes:
        correct = 0
        for i,j in zip(guess,outcome):
            if i == j: 
                correct += 1
            if correct == 2: 
                break
        if correct >= 2:
            result += 1
        else: 
            incorrect_outcomes.append(outcome)
    if result > best_result:
        best_result = result
        best_guess = guess
        best_incorrect = set(incorrect_outcomes)
print('outcomes:', len(outcomes))
print('number correct:', best_result)
print('best guess:',best_guess)
print('% outcomes >= 2 correct:', round(best_result / len(outcomes),2))
print('incorrect outcomes:')
for i in best_incorrect: print(i)
```
