```python
##### import matplotlib.pyplot as plt
import matplotlib

circle1 = plt.Circle((0, 0), 2, color='black', fill=False, alpha = 1)
circle2 = plt.Circle((3**(1/2), 0), 1, color='r', fill=False, alpha = 1) ## min position
circle3 = plt.Circle((2, 0), 1, color='blue', fill=False, alpha = 1) ## min position
fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()

ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)

ax.vlines(2,-3,3,color='blue',linestyle='--')
ax.vlines(3**(1/2),-3,3,color='red',linestyle='--')

ax.set_aspect(aspect=1);

plt.xlim(-3,3)
plt.ylim(-3,3)

plt.show();
```

```python
# circle formula
# (x−a)**2 + (y−b)**2 = r**2
# circle center is (a,b)
```

```python
from random import random
```

```python
# radius of outer circle = 1
# radius of inner disc = .5
# select 2 random floats between 0 & 1 as x and y position of disc center where (x^2 + y^2)**(1/2) <= 1
trials = 10000000
successes = 0
for i in range(trials):
    r = (random()**2 + random()**2)**(1/2)
    if r > 1:
        trials -= 1
        continue
    if r >= 3**(1/2)/2: 
        successes += 1 
print(round(successes/trials,4))
```
